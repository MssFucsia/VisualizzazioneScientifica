```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd

f = open("/workspace/Visualizzazione/NaturalDisasters.csv", "r")
lifeExp = []
cases = []
line = f.readline()
elements = line.split(";")
graphColor = "#f6f2ef"
plt.figure(facecolor = graphColor)
plt.axes(facecolor = graphColor)

    # Continua a leggere il file fino a quando ci sono righe
while line:
    elements = line.split(";")
    # Verifica se il primo elemento è "Central African Republic"
    if elements[0].strip() == "Chad":
        lifeExp.append(float(elements[1]))
        cases.append(int(float(elements[9])+0.5))

    # Leggi la riga successiva
    line = f.readline()
    
f.close()
years = []
for i in range (1980,2022,1):
    years.append(i)

df = pd.DataFrame({'y1': lifeExp, 'y2': cases, 'years': years })

plt.title("Natural Disasters in\nChad")
plt.xlabel("\nyears", loc='right')
plt.ylim(0, 60)
plt.yticks([0, 10, 20, 30, 40, 50, 60], labels = ['', '10', '20', '30', '40', '50', '60'])
plt.ylabel("age", rotation='horizontal', loc = 'top')
df['y1'].plot(kind='line')

df['y2'].plot(kind='bar', color='orange', secondary_y=True)
plt.ylim(0, 80)

# etichetta asse y2
plt.text(59, 82, "population percentage", rotation='horizontal', ha='right', va='center')
plt.yticks([0, 20, 40, 60, 80], labels = ['', '20%', '40%', '60%', '80%'])

# asse x
plt.xticks(ticks=[0, 10, 20, 30, 40], labels =['', '', '', '', ''])
plt.text(0, -4, "1980", rotation='horizontal', ha='center')
plt.text(10, -4, "1990", rotation='horizontal', ha='center')
plt.text(20, -4, "2000", rotation='horizontal', ha='center')
plt.text(30, -4, "2010", rotation='horizontal', ha='center')
plt.text(40, -4, "2020", rotation='horizontal', ha='center')

LE = mpatches.Patch(color='blue', label='Life Expectancy')
PP = mpatches.Patch(color='orange', label='Percentage of affected population')
plt.legend(handles=[LE, PP],ncol = 1, bbox_to_anchor=(1.12, 0.95), loc='upper left', facecolor = graphColor)

plt.show()
```
